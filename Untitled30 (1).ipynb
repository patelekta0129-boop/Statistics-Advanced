**Q1: What is a random variable in probability theory?**

**Ans:** A random variable is a function that assigns numerical values to the outcomes of a random experiment. It can be **discrete** (taking countable values, like the result of a dice roll) or **continuous** (taking values from an interval, like a person’s height).

**Q2: What are the types of random variables?**

**Ans:** **Types of Random Variables:**  
1. **Discrete Random Variable** – takes countable values, such as the outcome of a dice roll or number of heads in coin tosses.  
2. **Continuous Random Variable** – takes values from a continuous range, such as a person’s height or the time taken to complete a race.  

 In simple words: discrete = countable outcomes, continuous = measurable outcomes over an interval.

**Q3: Explain the difference between discrete and continuous distributions.**

**Ans:** **Difference between Discrete and Continuous Distributions:**  
- **Discrete Distribution**: Deals with random variables that take **countable values**. Example – probability of getting heads in coin tosses or rolling a die. Represented by a **Probability Mass Function (PMF)**.  
- **Continuous Distribution**: Deals with random variables that take values from a **continuous range** (uncountably infinite). Example – height of people or time taken to complete a race. Represented by a **Probability Density Function (PDF)**.  



**Q4: What is a binomial distribution, and how is it used in probability?**

**Ans:** **Binomial Distribution:**  
A binomial distribution is a discrete probability distribution that gives the likelihood of getting a certain number of successes in a fixed number of independent trials, where each trial has only two possible outcomes: success or failure.  

**Use in Probability:**  
It is used to model situations like coin tosses, quality control (defective vs non‑defective items), medical trials (success vs failure of treatment), and other experiments where outcomes are binary.  



 **Q5:What is the standard normal distribution, and why is it important?**

 **Ans:** **Standard Normal Distribution:**  
The standard normal distribution is a special case of the normal distribution with a **mean of 0** and a **standard deviation of 1**. It is symmetric and bell‑shaped, centered at zero.  

**Importance:**  
- It simplifies probability calculations because values can be converted into **z‑scores**.  
- Many statistical methods (like hypothesis testing and confidence intervals) rely on it.  
- It serves as a reference for comparing other normal distributions.  

 **Q6:What is the Central Limit Theorem (CLT), and why is it critical in statistics?**

 **Ans:** **Central Limit Theorem (CLT):**  
The Central Limit Theorem states that when independent random samples are taken from any population, the distribution of their sample mean tends to become approximately **normal (bell‑shaped)** as the sample size increases, regardless of the population’s original distribution.  

**Why It Is Critical in Statistics:**  
- It allows statisticians to use the **normal distribution** as an approximation for many problems.  
- It is the foundation for **hypothesis testing** and **confidence intervals**.  
- It makes analyzing data easier because even if the population is not normal, the sample mean will be nearly normal for large samples.



 **Q7:What is the significance of confidence intervals in statistical analysis?**

 **Ans:** **Confidence Intervals in Statistical Analysis:**  
A confidence interval gives a range of values within which the true population parameter (like mean or proportion) is expected to lie, with a certain level of confidence (commonly 95% or 99%).  

**Significance:**  
- It shows the **precision** of an estimate (narrow intervals = more precise).  
- It provides a measure of **uncertainty** instead of relying on a single point estimate.  
- It is widely used in **hypothesis testing**, **decision‑making**, and interpreting data reliability.

**Q8: What is the concept of expected value in a probability distribution?**

**Ans:** **Expected Value in Probability Distribution:**  
The expected value (also called the mean) is the long‑run average outcome of a random variable. It represents the value you would expect if an experiment were repeated many times.  

- For a **discrete random variable**:  
\[
E[X] = \sum x \cdot P(X = x)
\]  
(sum of each possible value multiplied by its probability)  

- For a **continuous random variable**:  
\[
E[X] = \int_{-\infty}^{\infty} x \cdot f(x) \, dx
\]  
(integral of the variable times its probability density function).  

**Significance:**  
- It gives the central tendency of a probability distribution.  
- Used in decision‑making, risk analysis, economics, and statistics.  
- Helps predict the average outcome in the long run.

 **Q8: Write a Python program to generate 1000 random numbers from a normal
distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
deviation using NumPy, and draw a histogram to visualize the distribution.**

**Ans:**


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random numbers from normal distribution (mean=50, std=5)
data = np.random.normal(loc=50, scale=5, size=1000)

# Compute mean and standard deviation
mean_val = np.mean(data)
std_val = np.std(data)

print("Computed Mean:", mean_val)
print("Computed Standard Deviation:", std_val)

# Plot histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Normal Distribution (mean=50, std=5)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

**Q10 :You are working as a data analyst for a retail company. The company has
collected daily sales data for 2 years and wants you to identify the overall sales trend.
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
● Explain how you would apply the Central Limit Theorem to estimate the average sales
with a 95% confidence interval.
● Write the Python code to compute the mean sales and its confidence interval.**

**Ans:** The **Central Limit Theorem (CLT)** states that if we take repeated random samples from a population, the distribution of the sample means will approach a normal distribution as the sample size increases, regardless of the population’s original distribution.  

In this case, the company’s **daily sales data** can be treated as samples. By applying CLT, we can assume the sample mean follows a normal distribution. Using this, we can construct a **95% confidence interval (CI)** for the true average daily sales. The CI is calculated as:

```python
import numpy as np
import scipy.stats as st

# Daily sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Convert to numpy array
data = np.array(daily_sales)

# Compute mean and standard deviation
mean_sales = np.mean(data)
std_sales = np.std(data, ddof=1)  # sample standard deviation
n = len(data)

# 95% confidence interval using CLT
confidence_level = 0.95
alpha = 1 - confidence_level
z_score = st.norm.ppf(1 - alpha/2)

margin_of_error = z_score * (std_sales / np.sqrt(n))
lower_bound = mean_sales - margin_of_error
upper_bound = mean_sales + margin_of_error

print("Mean Sales:", mean_sales)
print("Standard Deviation:", std_sales)
print("95% Confidence Interval: ({:.2f}, {:.2f})".format(lower_bound, upper_bound))
```
